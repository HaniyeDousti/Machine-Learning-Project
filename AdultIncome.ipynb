```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, KFold, cross_val_score, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
```

```python
adult_dataset = pd.read_csv('adult.csv')
```

```python
adult_dataset.isnull().sum()
```

```python
# Convert sex value to 0 and 1 
# Male -> 0 and female -> 1
adult_dataset["gender"] = adult_dataset["gender"].map({"Male": 0, "Female": 1})

# Create Married Column - Binary Yes(1) or No(0)
adult_dataset['marital-status'] = adult_dataset['marital-status'].replace(['Never-married', 'Divorced', 'Separated', 'Widowed'], 0)
adult_dataset['marital-status'] = adult_dataset['marital-status'].replace(['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse'], 1)

adult_dataset['marital-status'] = adult_dataset['marital-status'].astype(int)
```

```python
adult_dataset.info()
```

```python
adult_dataset.head(10)
```

```python
adult_dataset.drop(labels=['workclass', 'education', 'relationship', 'race', 'native-country', 'occupation'], axis=1, inplace=True)
adult_dataset.head()
```

```python
y = adult_dataset['income']
adult_dataset = adult_dataset.drop('income', axis=1)
```

**Splitting the dataset into train and test data**

```python
X_train, X_test, y_train, y_test = train_test_split(adult_dataset, y, test_size=0.2, random_state=42)
```

**Training Model**

```python
models = []
models.append(('LR', LogisticRegression()))
models.append(('KMN', KNeighborsClassifier()))
print(models)
```

```python
results = dict()
for name, model in models:
  Kfold = KFold(n_splits=10, shuffle= True, random_state=7)
  CV_results = cross_val_score(model, X_train, y_train, cv=Kfold, scoring='accuracy', error_score='raise')
  results[name] = (CV_results.mean(), CV_results.std())



print()
print('name    results.mean    results.std')
for key, value in results.items():
  print(key, value)
```
