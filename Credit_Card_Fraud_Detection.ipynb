```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading the dataset
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
#Checking the number of missing value in each column
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

This dataset is unbalance because 99% of the dataset is legit transaction. We need to do some processing to make it balance.

0 --> Legit Transaction

1 --> Fraudulent Transaction

**Seperating the data for analysis**

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

**Under-Sampling**

Building a sample dataset containing similar distribution of legit and fraudulent transactions

```python
legit_sample = legit.sample(n=492)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

So, the dataset is balance now

```python
new_dataset.groupby('Class').mean()
```

```python
new_dataset.describe()
```

```python
legit2 = new_dataset[new_dataset.Class == 0]
fraud2 = new_dataset[new_dataset.Class == 1]
```

```python
legit2.Amount.describe()
```

```python
fraud2.Amount.describe()
```

**Splitting the data into features and target**

```python
X = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

**Splitting the data into training and test data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

**Training the model**

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

**Evaluation**

Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
training_data_accuracy
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
test_data_accuracy
```
